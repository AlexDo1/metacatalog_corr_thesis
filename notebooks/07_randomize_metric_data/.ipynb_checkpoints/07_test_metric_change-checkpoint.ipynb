```python
from metacatalog_corr.models import CorrelationMetric

# load data
left = api.find_entry(session, id=7)[0]
print(left.title)
left = left.get_data()
right = api.find_entry(session, id=4)[0]
print(right.title)
right=right.get_data()

# harmonize data by temporal indices
left, right = thesis_functions.harmonize_data(left, right)

fig, ax = plt.subplot_mosaic(
    '''
    AB
    CC
    ''', figsize=[15,15])

# plot original data
sns.scatterplot(x=left, y=right, ax=ax['A'])
sns.scatterplot(x=left, y=right, ax=ax['B'])


metric = 'pearson'

# get metric object
metric = session.query(CorrelationMetric).filter(CorrelationMetric.symbol == metric).one()

# calculate metric on original left and right data
corr_orig = metric.calc(left, right)

# list to store metric values from left and right with replacements leading to lower correlation
corr_lower_corr = []

right_lower_corr = right.copy()

# generate index array
right_idx_array = np.linspace(start=0, stop=len(right), num=len(right), dtype=int, endpoint=False)
# shuffle index array
np.random.shuffle(right_idx_array)

for n, right_idx in enumerate(right_idx_array):
    # replace value in right at random right_index with random number from normal distribution with same mean and std as original right data
    right_lower_corr[right_idx] = np.random.default_rng().normal(loc=right.mean(), scale=right.std())
    
    # recalculate metric with altered right data
    corr_replace = metric.calc(left, right_lower_corr)
    
    # append to corr_replace_list
    corr_lower_corr.append(corr_replace)
    
    
# plot altered data
sns.scatterplot(x=left, y=right_lower_corr, alpha=0.2, ax=ax['A'])






# list to store metric values from left and right with replacements leading to higher correlation
corr_higher_corr = []

right_higher_corr = right.copy()

# generate index array
right_idx_array = np.linspace(start=0, stop=len(right), num=len(right), dtype=int, endpoint=False)
# shuffle index array
np.random.shuffle(right_idx_array)

for n, right_idx in enumerate(right_idx_array):
    # replace value in right at random right_index with random number from normal distribution with same mean and std as original right data
    right_higher_corr[right_idx] = left[right_idx]
    
    # recalculate metric with altered right data
    corr_replace = metric.calc(left, right_higher_corr)
    
    # append to corr_replace_list
    corr_higher_corr.append(corr_replace)
    
    
# plot altered data
sns.scatterplot(x=left, y=right_higher_corr, alpha=0.2, ax=ax['B'])


# plot metric
sns.lineplot(x=[-1*x for x in list(range(len(right)))], y=corr_higher_corr, ax=ax['C'])
sns.lineplot(x=range(0,len(right)), y=corr_lower_corr, ax=ax['C'])
ax['C'].scatter(x=0, y=corr_orig, marker='o', s=100, c='r')
```

```python
from metacatalog_corr.models import CorrelationMetric
from skinfo.metrics import mutual_information, get_2D_bins

# load data
left = api.find_entry(session, id=7)[0]
print(left.title)
left = left.get_data()
right = api.find_entry(session, id=4)[0]
print(right.title)
right=right.get_data()

# harmonize data by temporal indices
left, right = thesis_functions.harmonize_data(left, right)

fig, ax = plt.subplot_mosaic(
    '''
    AB
    CC
    ''', figsize=[15,15])

# plot original data
sns.scatterplot(x=left, y=right, ax=ax['A'])
sns.scatterplot(x=left, y=right, ax=ax['B'])


metric = 'mutual_info'

# get metric object
#metric = session.query(CorrelationMetric).filter(CorrelationMetric.symbol == metric).one()
bins = get_2D_bins(left, right, bins='fd', same_bins=True)


# calculate metric on original left and right data
corr_orig = mutual_information(left, right, bins=bins, normalize=True)

# list to store metric values from left and right with replacements leading to lower correlation
corr_lower_corr = []

right_lower_corr = right.copy()

# generate index array
right_idx_array = np.linspace(start=0, stop=len(right), num=len(right), dtype=int, endpoint=False)
# shuffle index array
np.random.shuffle(right_idx_array)

for n, right_idx in enumerate(right_idx_array):
    # replace value in right at random right_index with random number from normal distribution with same mean and std as original right data
    right_lower_corr[right_idx] = np.random.default_rng().normal(loc=right.mean(), scale=right.std())
    
    # recalculate metric with altered right data
    corr_replace = mutual_information(left, right_lower_corr, bins=bins, normalize=True)
    
    # append to corr_replace_list
    corr_lower_corr.append(corr_replace)
    
    
# plot altered data
sns.scatterplot(x=left, y=right_lower_corr, alpha=0.2, ax=ax['A'])






# list to store metric values from left and right with replacements leading to higher correlation
corr_higher_corr = []

right_higher_corr = right.copy()

# generate index array
right_idx_array = np.linspace(start=0, stop=len(right), num=len(right), dtype=int, endpoint=False)
# shuffle index array
np.random.shuffle(right_idx_array)

for n, right_idx in enumerate(right_idx_array):
    # replace value in right at random right_index with random number from normal distribution with same mean and std as original right data
    right_higher_corr[right_idx] = left[right_idx]
    
    # recalculate metric with altered right data
    corr_replace = mutual_information(left, right_higher_corr, bins=bins, normalize=True)
    
    # append to corr_replace_list
    corr_higher_corr.append(corr_replace)
    
    
# plot altered data
sns.scatterplot(x=left, y=right_higher_corr, alpha=0.2, ax=ax['B'])


# plot metric
sns.lineplot(x=[-1*x for x in list(range(len(right)))], y=corr_higher_corr, ax=ax['C'])
sns.lineplot(x=range(0,len(right)), y=corr_lower_corr, ax=ax['C'])
ax['C'].scatter(x=0, y=corr_orig, marker='o', s=100, c='r')
```
