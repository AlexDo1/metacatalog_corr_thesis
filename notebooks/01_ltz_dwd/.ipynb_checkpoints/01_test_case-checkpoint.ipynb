Dendrogram Plot

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(clustering, truncate_mode='level', p=1)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.tight_layout()
plt.show()
```

### Reorganize data
table: left_id, right_id, all metrics

```python
left_ids = ['8', '9']
left_ids = ', '.join(left_ids)
right_ids = left_ids

sql = 'SELECT * FROM correlation_matrix WHERE left_id IN (%s) AND right_id IN (%s)' % (left_ids, right_ids)
df = pd.read_sql(sql, session.bind)
df
```

```python
def reorganize_matrix(session, left_entries, right_entries):
    """
    Function to transform the correlation_matrix in metacatalog to a wide format with columns
    left_id, right_id and all metric values.
    
    Parameters
    ----------
    session : sqlalchemy.orm.Session
        Session to the database.
    left_entries : list(metacatalog.models.Entry)
        List of entries for which the correlation matrix is to be loaded.
    right_entries : list(metacatalog.models.Entry)
        List of entries for which the correlation matrix is to be loaded.
    """
    # get left_id's
    left_ids = []
    for entry in left_entries:
        left_ids.append(str(entry.id))
    left_ids = ', '.join(left_ids) # remove brackets from list of ids to use them in the sql query below
    
    # get right_id's
    right_ids = []
    for entry in right_entries:
        right_ids.append(str(entry.id))
    right_ids = ', '.join(right_ids) # remove brackets from list of ids to use them in the sql query below
    
    
    # load correlation matrix for specified left_id's and right_id's
    sql = 'SELECT * FROM correlation_matrix WHERE left_id IN (%s) AND right_id IN (%s)' % (left_ids, right_ids)
    df = pd.read_sql(sql, session.bind)
    
    # get metrics for output column names
    sql = 'SELECT id, symbol FROM correlation_metrics'
    df_metrics = pd.read_sql(sql, session.bind)
    dict_metrics = dict(zip(df_metrics.id, df_metrics.symbol))
    
    # build result_dict, convert to pd.DataFrame later
    result_dict = {'left_id': [],
                   'right_id': []}
    
    # key for each metric
    for metric_id in sorted(df.metric_id.unique()):
        result_dict[dict_metrics[metric_id]] = []

    # fill dictionary
    for left_id in sorted(df.left_id.unique()):    
        for right_id in sorted(df.right_id.unique()):       
            result_dict['left_id'].append(left_id)            
            result_dict['right_id'].append(right_id)            
            for metric_id in sorted(df.metric_id.unique()):
                metric_value = df.loc[((df['left_id']==left_id) & (df['right_id']==right_id) & (df['metric_id']==metric_id))]['value']
                # add value if existing
                if metric_value.any():
                    result_dict[dict_metrics[metric_id]].append(float(metric_value))
                else:
                    result_dict[dict_metrics[metric_id]].append(np.nan)
    # create df from dict
    result_df = pd.DataFrame(result_dict)
    
    return result_df
```

Build function: 
- **input**: two datasets (left, right), metrics2compute
- **output**: timeseries plot, scatter plot, qq-plots, table with all metrics

```python
import pandas as pd
import numpy as np
from matplotlib.dates import DateFormatter, MonthLocator
import statsmodels.api as sm


def calculate_metrics_set(left, right, corr_metrics):
    """
    Calculate a set of correlation metrics for 1D timeseries DataFrames left and right.
    Also plots timeseries and scatter plot with linear regression.
    
    Parameters
    ----------
    left : pandas.DataFrame
        Left 1D timeseries DataFrame for which the metrics are calculated.
    right : pandas.DataFrame
        Right 1D timeseries DataFrame for which the metrics are calculated.
    corr_metrics : list[str]
        List of metrics to compute ('spearman', 'pearson').
        
    Returns
    -------
    df : pandas.DataFrame
        DataFrame with the values of the calculated metrics.
    """
    
    # append calculated metrics to data
    data = []
    
    if 'pearson' in corr_metrics:
        pearson_r = metrics.pearson_corr_coef(left.iloc[:,0], right.iloc[:,0])
        data.append({'metric': 'pearson_r', 
                     'value': round(pearson_r, 3)})
        
    if 'spearman' in corr_metrics:
        spearman_r = metrics.spearman_corr_coef(left.iloc[:,0], right.iloc[:,0])
        data.append({'metric': 'spearman_r', 
                     'value': round(spearman_r, 3)})
    
    # save metric results to df
    df = pd.DataFrame(data)
    
    # timeseries and scatter plot with linear regression & table with calculated metrics
    #fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,6))
    fig = plt.figure(figsize=(12,6))

    gs = fig.add_gridspec(2, 3)
    
    ax1 = fig.add_subplot(gs[0:2, 0])
    
    ax2 = fig.add_subplot(gs[0:2, 1])
    
    ax3 = fig.add_subplot(gs[0, 2])
    
    ax4 = fig.add_subplot(gs[1, 2])
    
    # timeseries plot
    # secondary y-axis
    ax1b = ax1.twinx()
    
    # plotting on the first y-axis
    ax1.plot('%s' % left.columns[0], data=left, alpha=1, color='slateblue')
    ax1.set(xlabel='Date',
            ylabel='%s' % left.columns[0], # not needed, scatterplot with same y-axis label
            title='Timeseries')
    
    # plotting on the second y-axis
    ax1b.set_ylabel('%s' % right.columns[0])
    ax1b.plot('%s' % right.columns[0], data=right, alpha=0.4, color='darkturquoise')
    
    # lines and labels from all axes for a single legend
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax1b.get_legend_handles_labels()
    ax1b.legend(lines + lines2, labels + labels2, loc='upper left');
    
    # format date labels
    plt.setp(ax1.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    
    # scatter plot with linear regression line    
    # linear regression
    x = left.iloc[:,0]
    y = right.iloc[:,0]
    coef = np.polyfit(x,y,1)
    poly1d_fn = np.poly1d(coef) 
    # poly1d_fn is now a function which takes in x and returns an estimate for y

    ax2.scatter(x, y, color='teal')
    ax2.plot(x, poly1d_fn(x), '--k')
    ax2.set(xlabel='%s' % left.columns[0],
            #ylabel='%s' % right.columns[0], # not needed -> timeseries plot with same y-label next to scatterplot
            title="Linearity?")
    
    
    # qq-plots for dat1 and dat2
    sm.qqplot(dat1.iloc[:,0], line='q', ax=ax3)
    ax3.set(title='Normality %s?' % dat1.columns[0])
    
    sm.qqplot(dat2.iloc[:,0], line='q', ax=ax4)
    ax4.set(title='Normality %s?' % dat2.columns[0])
    
    
    fig.tight_layout()
    
    plt.show()
    
    return df    
    
```
