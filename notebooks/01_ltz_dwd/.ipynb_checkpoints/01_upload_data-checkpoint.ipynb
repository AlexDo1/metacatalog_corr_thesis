DWD: Die Globalstrahlung umfasst den direkten und diffusen Anteil der solaren Strahlung bezogen auf die Horizontalfläche. Manchmal wird "Globalstrahlung" auch mit der Bezeichnung "kurzwellig" verknüpft, dabei ist bis zu 2.8 Mikrometer gemeint, denn in diesem Zusammenhang bezieht sich "kurzwellig" auf das solare Spektrum, im Gegensatz zu "langwellig" als Bezeichnung des Spektrums der Wärmestrahlung der Atmosphäre.

#### global_radiation == solar_irradiance?:
- LTZ unit: **[Wh/m2]** -> stündliche Auflösung == **[W/m2]**
$$LTZ: \frac{Wh}{m^2} = 1h * \frac{W}{m^2}$$
- DWD unit: **[J/cm2]** -> 10min Messzeitraum (600s):
$$DWD: \frac{1}{600 s}\frac{J}{cm^2} = \frac{10000}{600} \frac{W}{m^2}$$
- metacatalog: 'solar irradiance' **[W/m2]**

**factor 10000/resolution[s] to convert DWD data from [J/cm2] to [W/m2]**

```python
UPLOAD = True

# Local
DATA = './data'
CONNECTION = 'mc_corr'
```

```python
session = api.connect_database(CONNECTION)
print(session.bind)
```

# Load LTZ Augustenberg data

```python
df = pd.read_csv('data/FOR20210730-093332S.txt', sep='\t', decimal=',', skiprows=4, encoding = 'ISO-8859-1')
# directly name columns like corresponding variables in metacatalog
df.columns = ['tstamp', 'hour', 'soil temperature', 'global radiation', 'relative humidity', 'air temperature', 'precipitation', 'wind speed', 'leaf wetness']

# drop leaf wetness data for now -> not possible to store binary data in metacatalog atm (?)
df.drop('leaf wetness', axis=1, inplace=True)

df['tstamp'] = pd.to_datetime(df['tstamp'], format='%Y%m%d') + pd.to_timedelta(df.pop('hour'), unit='H')
df.set_index('tstamp', inplace=True)

df.head(2)
```

#### Variables

```python
# missing variables:
var_precipitation =  api.find_variable(session, name='precipitation', return_iterator=True).first()
if var_precipitation is None and UPLOAD:
    var_precipitation = api.add_variable(session, name='precipitation', symbol='P', column_names=['precipitation'], unit=102)

var_wind_speed = api.find_variable(session, name='wind speed', return_iterator=True).first()
if var_wind_speed is None and UPLOAD:
    var_wind_speed = api.add_variable(session, name='wind speed', symbol='u', column_names=['wind_speed'], unit=107)

var_radiation = api.find_variable(session, name='global radiation', return_iterator=True).first()
if var_radiation is None and UPLOAD:
    var_radiation = api.add_variable(session, name='global radiation', symbol='G', column_names=['global_radiation'], unit=115)

# do not import binary leaf wetness data at the moment
#var_leaf_wetness = api.find_variable(session, name='leaf wetness', return_iterator=True).first()
#unit_lw = api.find_unit(session, name='binary')
#if var_leaf_wetness is None and unit_lw is None and UPLOAD:
#    unit_lw = api.add_unit(session, name='binary', symbol='b01', si='1')
#    var_leaf_wetness = api.add_variable(session, name='leaf wetness', symbol='LW', column_names=['leaf_wetness'], unit=unit_lw.id)
```

#### Details: measuring height (from raw column names)

```python
details_dict = {'measuring_height': ['-5cm', np.nan, '2m', '2m', np.nan, '2.5m', np.nan]}
details_dict
```

#### Author / Owner

```python
# check if existing
author_ltz = api.find_person(session, organisation_name='Landwirtschaftliches Technologiezentrum Augustenberg', return_iterator=True).first()

if author_ltz is None and UPLOAD:
    author_ltz = api.add_person(session, first_name='', last_name='',
                                organisation_name='Landwirtschaftliches Technologiezentrum Augustenberg',
                                affiliation = 'Landwirtschaftliches Technologiezentrum Augustenberg, Landesanstalt für Pflanzenbau, Rheinstetten-Forchheim',
                                organisation_abbrev='LTZ'
                               )
print(author_ltz)
```

#### Create an Entry for each variable, upload corresponding data

```python
for idx, col in enumerate(df.columns):
    # add entry if not existing
    entry = api.find_entry(session, title='LTZ Augustenberg: ' + col, return_iterator=True).first()

    if not entry and UPLOAD:
        entry = api.add_entry(session, 
                              title='LTZ Augustenberg: ' + col, 
                              author=author_ltz.id, 
                              location=(8.3357, 48.9659),
                              variable=col, 
                              license=6, 
                              embargo=True)
        
        # add details
        if isinstance(details_dict['measuring_height'][idx], str):
            details = {
                'key': 'measuring_height',
                'value': details_dict['measuring_height'][idx],   
                'description': 'measuring height)'
            }
            
        api.add_details_to_entries(session, [entry], details=[details])
         
    # create datasource, add datasource scale
    if not entry.datasource and UPLOAD:
        entry.create_datasource(type=1, path='timeseries', datatype='timeseries', commit=True)
        entry.datasource.create_scale(resolution='1hour', 
                                      extent=(df.index[0], df.index[-1]), 
                                      support=1.0, # not sure 
                                      scale_dimension='temporal')
        # add data
        entry.import_data(df[col])
        
        session.commit()
        
        print('%d/%d: %s data loaded.' % (idx+1, len(df.columns), col))
```

#### EntryGroup

```python
group_ltz = api.find_group(session, title='LTZ Augustenberg', return_iterator=True).first()

if not group_ltz and UPLOAD:
    entries = api.find_entry(session, title='LTZ Augustenberg*')
    group_ltz = api.add_group(session, 'Project',
                          entry_ids=[e.id for e in entries],
                          title='LTZ Augustenberg',
                          description='Meteorological data from LTZ Augustenberg used to test metrics in metacatalog_corr.'
                         )
    
for e in group_ltz.entries:
    print(e.title)
```

#### Check data

```python
fig, axs = plt.subplots(1,6)

m = 0

for e in group_ltz.entries:
    edat = e.get_data()
    edat.plot(title=e.variable.name + '_' + e.variable.unit.symbol, legend=False, ax=axs[m], figsize=(15,6))
    m += 1

plt.tight_layout()
plt.show()
```

# Load DWD data

#### Author / Owner

```python
# check if existing
author_dwd = api.find_person(session, organisation_name='Deutscher Wetterdienst', return_iterator=True).first()

if author_dwd is None and UPLOAD:
    author_dwd = api.add_person(session, first_name='', last_name='',
                                organisation_name='Deutscher Wetterdienst',
                                organisation_abbrev='DWD'
                               )
print(author_dwd)
```

We use the python package wetterdienst to load DWD data.  
License: https://wetterdienst.readthedocs.io/en/latest/data/license.html

Set environment variable WD_CACHE_DISABLE to TRUE to turn off caching when loading wetterdienst data.

```python
%env WD_CACHE_DISABLE=1
```

#### Choose parameters  
We first check if the same data is available as in the LTZ dataset:
- temperature_soil_005
- radiation_global
- humidity
- temperature_air_200
- precipitation_height
- wind_speed  

There are many more parameters available, including nominal data like cloud_type.  

```python
from wetterdienst.provider.dwd.observation import DwdObservationRequest
import numpy as np
import pandas as pd

# parameter_name_dwd: [variable.id, details(measuring height/depth)]
# further measurement heights probably in meta_data.zip: https://opendata.dwd.de/climate_environment/CDC/observations_germany/climate/1_minute/precipitation/
dwd_var_dict = {
    'temperature_soil_005': ['soil temperature', '-5cm'], 
    'radiation_global': ['global radiation', np.nan],
    'humidity': ['relative humidity', np.nan],
    'temperature_air_200': ['air temperature', '2m'],
    'precipitation_height': ['precipitation', np.nan],
    'wind_speed': ['wind speed', np.nan]
}

# resolution_name_dwd: datasource.create_scale_value [min], resolution [s] 
dwd_res_dict = {
    'minute_1': ['1min', 60],
    'minute_10': ['10min', 600],
    'hourly': ['60min', 3600],
    'subdaily': ['360min', 21600],
    'daily': ['1440min', 86400],
    'monthly': ['43200min', 2592000],
    'annual': ['15768000min', 946080000]
}

def import_into_metacatalog(var_list, res_list):
    """
    Function to create entries and load data for every available 
    combination of variable in var_list and resolution in res_list.
    
    Parameters
    ----------
    var_list : list[str]
        List of variables to download data from DWD server and import to metacatalog.
    res_list : list[str]
        List of resolutions to download data from DWD server and import to metacatalog.
    """
    for var in var_list:
        for res in res_list:
            request = DwdObservationRequest(parameter=var,
                                            resolution=res,
                                            start_date='2000-01-01',  # if not given timezone defaulted to UTC
                                            end_date='2020-01-01',  # if not given timezone defaulted to UTC
                                            tidy=True,  # default, tidy data
                                            humanize=True,  # default, humanized parameters
                                            si_units=False  # default, convert values to SI units
                                           ).filter_by_name('Rheinstetten')
            
            # if request is not empty: data for variable - resolution combination exists 
            if not request.df.empty:
                try:
                    #----- load data from DWD server
                    df = request.values.all().df.dropna()  
                    
                    # drop columns
                    df.drop(['station_id', 'dataset', 'parameter', 'quality'], axis=1, inplace=True)
                    # datetime format
                    df.date = df.date.astype(str)
                    df.date.replace('\+00:00', '', regex=True, inplace=True)
                    df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
                    # set tstamp index
                    df.set_index('date', inplace=True)
                    df.index.rename('tstamp', inplace=True)
                    # drop duplicates in index
                    df = df[~df.index.duplicated(keep='first')]

                    # rename data column
                    df.columns = [var]
                    
                    # data conversion: solar irradiance [J/cm2] -> [W/m2]
                    if var == 'radiation_global':
                        df[var] = df[var] * (10000/dwd_res_dict[res][1])
                    
                    #----- Upload to metacatalog
                    # add entry if not existing
                    entry = api.find_entry(session, title='DWD station Rheinstetten, %s, resolution: %s' % (dwd_var_dict[var][0], res), return_iterator=True).first()
        
                    if not entry and UPLOAD:
                        entry = api.add_entry(session, 
                                              title='DWD station Rheinstetten, %s, resolution: %s' % (dwd_var_dict[var][0], res), 
                                              author=author_dwd.id, 
                                              abstract='%s data with resolution "%s" from the DWD meteorological measuring station Rheinstetten, Quelle: Deutscher Wetterdienst' % (var.capitalize(), res),
                                              location=(8.3301, 48.9726),
                                              variable=dwd_var_dict[var][0], 
                                              license=6, 
                                              embargo=True)
                                    
                        # add details
                        if isinstance(dwd_var_dict[var][1], str):
                            details = {
                                'key': 'measuring_height',
                                'value': dwd_var_dict[var][1],   
                                'description': 'measuring height'
                            }
                            
                            api.add_details_to_entries(session, [entry], details=[details])
                    
                    # create datasource, add datasource scale
                    if not entry.datasource and UPLOAD:
                        entry.create_datasource(type=1, path='timeseries', datatype='timeseries', commit=True)
                        entry.datasource.create_scale(resolution='%s' % dwd_res_dict[res][0], 
                                                      extent=(df.index[0], df.index[-1]), 
                                                      support=1.0, # not sure 
                                                      scale_dimension='temporal')
                        # add data
                        entry.import_data(df)
                        
                        session.commit()
                        
                        print('%s %s data loaded' % (dwd_var_dict[var][0], res))
                except KeyError as e:
                    # ???
                    print('[KeyError]: %s, %s: %s' % (var, res, str(e)))
                except ValueError as e:
                    # ???
                    print('[ValueError]: %s, %s: %s' % (var, res, str(e)))
    
```

```python
var_list = ['temperature_soil_005', 'radiation_global', 'humidity', 'temperature_air_200', 'precipitation_height', 'wind_speed']
res_list = ['minute_1','minute_10','hourly','subdaily','daily','monthly','annual']

import_into_metacatalog(var_list, res_list)
```

#### EntryGroup

```python
group_dwd = api.find_group(session, title='DWD station Rheinstetten', return_iterator=True).first()

if not group_dwd and UPLOAD:
    entries = api.find_entry(session, title='DWD station Rheinstetten*')
    group_dwd = api.add_group(session, 'Project',
                              entry_ids=[e.id for e in entries],
                              title='DWD station Rheinstetten',
                              description='Meteorological data from DWD used to test metrics in metacatalog_corr.'
                             )
for e in group_dwd.entries:
    print(e.title)
```

#### Check data

```python
fig, axs = plt.subplots(6,4)

m = 0
n = 0
o = 0
p = 0
q = 0
r = 0

for e in group_dwd.entries:
    if 'soil temperature' in e.title:
        edat = e.get_data()
        edat.plot(title=e.title[26:], legend=False, color='r', ax=axs[0, m], figsize=(15,10))
        m += 1
    if 'global radiation' in e.title:
        edat = e.get_data()
        edat=edat*10000
        edat.plot(title=e.title[26:], legend=False, color='y', ax=axs[1, n], figsize=(15,10))
        n += 1
    if 'relative humidity' in e.title:
        edat = e.get_data()
        edat.plot(title=e.title[26:], legend=False, color='b', ax=axs[2, o], figsize=(15,10))
        o += 1
    if 'air temperature' in e.title:
        edat = e.get_data()
        edat.plot(title=e.title[26:], legend=False, color='g', ax=axs[3, p], figsize=(15,10))
        p += 1
    if 'precipitation' in e.title:
        edat = e.get_data()
        edat.plot(title=e.title[26:], legend=False, color='c', ax=axs[4, q], figsize=(15,10))
        q += 1    
    if 'wind speed' in e.title:
        edat = e.get_data()
        edat.plot(title=e.title[26:], legend=False, color='m', ax=axs[5, r], figsize=(15,10))
        r += 1

plt.tight_layout()

plt.show()
```
