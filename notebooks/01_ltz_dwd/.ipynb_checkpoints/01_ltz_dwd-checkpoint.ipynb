## p-value analysis

```python
# get table with p-values
sql=f"select * from correlation_matrix where identifier like '%%FINAL%%' and left_id in ({', '.join(list(map(str, id_ltz_dwd)))}) and right_id in ({', '.join(list(map(str, id_ltz_dwd)))}) order by left_id, right_id, metric_id"
df_pval = pd.read_sql(sql, session.bind)

# filter for LTZ and DWD entries only
df_pval = df_pval[df_pval.left_id.isin(id_ltz_dwd) & df_pval.right_id.isin(id_ltz_dwd)]

# rename metric_id to metric_name
sql = 'SELECT id, symbol FROM correlation_metrics'
df_metrics = pd.read_sql(sql, session.bind)
dict_metrics = dict(zip(df_metrics.id, df_metrics.symbol))
df_pval.metric_id.replace(dict_metrics, inplace=True)

df_pval
```

In the following cell, the occurence of p-values >= 0.05 is counted for every metric in certain ranges of metric scores.  
When the correlation score is already low, the p-value is not meaningful as we test for zero correlation and their is NO correlation indeed!

```python
count_dict = {}
df_pval_count = pd.DataFrame(columns=select_metrics)

bin_tuples = [(0,0.1), (0.1,0.3), (0.3,0.6), (0.6,0.8), (0.8,1)]

for bin_tuple in bin_tuples:
    for metric in select_metrics:
        df_sub = df_pval[(df_pval.p_value >= 0.05) & (df_pval.value >= bin_tuple[0]) & (df_pval.value <= bin_tuple[1])]
        count_dict[metric] = len(df_sub[df_sub.metric_id == metric])
    
    df_pval_count = df_pval_count.append(count_dict, ignore_index=True)

df_pval_count.index = list(map(str, bin_tuples))
df_pval_count.index.name = 'absolute correlation score range'

df_pval_count.to_csv('tables/ltz_dwd_p_value_range.csv')
df_pval_count
```

# 3.) Compare metrics: Clustering

```python
# split up variables in identifier into new columns left_variable, right_variable
for idx, row in df.iterrows():
    df.loc[idx, 'left_variable'] = df.loc[idx, 'identifier'].split('[')[1].split(']')[0].replace(',', '').split()[0]
    df.loc[idx, 'right_variable'] = df.loc[idx, 'identifier'].split('[')[1].split(']')[0].replace(',', '').split()[1]

# create keys column for variable combinations with expected correlation (see below)
for idx, row in df.iterrows():
    # sorted: easily get variable combinations in both directions
    df.loc[idx, 'var_combination_key'] = (str(sorted([row.left_variable, row.right_variable])))
```

```python
df.iloc[:,3:].groupby(df.var_combination_key).median()
```

#### add column expected correlation

reshape for LaTEX

| **Criterion**                                                  | **Points** |
|:---------------------------------------------------------------|-----------:|
| left_id = right_id                 |          1 |
| variable_combinations:<br>air temperature <-> soil temperature |        0.9 |
| left_variable = right_variable  (except precipitation)                                  |        0.8 |
| variable_combinations:<br>global radiation <-> soil temperature|        0.5 |
| variable_combinations:<br>relative_humidity <-> soil_temperature |          0.5 |
| variable_combinations:<br>air_temperature <-> global_radiation|          0.5 |
| variable_combinations:<br>global_radiation <-> relative_humidity|          0.5 |
| variable_combinations:<br>air_temperature <-> relative_humidity|          0.5 |

```python
# dictionary: may give different scores for different variable combinations
variable_combinations_strong = ["['air_temperature', 'soil_temperature']"]
variable_combinations_medium = ["['global_radiation', 'soil_temperature']",
                                "['relative_humidity', 'soil_temperature']",
                                "['air_temperature', 'global_radiation']",
                                "['global_radiation', 'relative_humidity']",
                                "['air_temperature', 'relative_humidity']"
                               ]
variable_exception = ["['precipitation', 'precipitation']"]




conditions = [
    df['left_id'].eq(df['right_id']), # 1
    df['var_combination_key'].isin(variable_exception), # 0
    df['left_variable'].eq(df['right_variable']), # 0.8
    df['var_combination_key'].isin(variable_combinations_strong), # 0.9
    df['var_combination_key'].isin(variable_combinations_medium) # 0.5
]

choices = [1, 0, 0.8, 0.9, 0.5]

# When multiple conditions are satisfied, the first one encountered in conditions is used.
df['expected_corr'] = np.select(conditions, choices, default=0)

# save dataframe
df.to_csv('df_clustering.csv')

df.head()
```

## Main Clustering cell for all metrics

```python
# for annotation
import matplotlib.transforms as mtransforms

for metric in metric_names:    
    df_cluster = df.copy()
    
    fig, ax = plt.subplots(1, figsize=(6, 3))
    
    # compute clustering, label_dict stores x-axis information
    clustering, label_dict = thesis_functions.clustering_and_plot_dendrogram(df_cluster, metric=metric, compare_to_pearson=False, ax=ax)
    
    # annotate a)
    trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
    plt.text(0, 1, 'a)', verticalalignment='top', transform = ax.transAxes+trans)
    
    # save dendrogram
    plt.tight_layout()
    plt.savefig(f'figures/Custering/{metric}/{metric}_dendrogram.pdf', dpi=100)
    plt.close()
    
    # column in df_cluster to store cluster number of each row
    df_cluster['Cluster'] = clustering.labels_
    
    # rename cluster labels (0, 1, 2, 3) in ascending order (by average corr value in cluster) to Cluster I, Cluster II, ...
    cluster_mean_dict = {}
    
    cluster_mean_dict[0] = df_cluster[metric][clustering.labels_==0].mean()
    cluster_mean_dict[1] = df_cluster[metric][clustering.labels_==1].mean()
    cluster_mean_dict[2] = df_cluster[metric][clustering.labels_==2].mean()
    cluster_mean_dict[3] = df_cluster[metric][clustering.labels_==3].mean()
    
    sorted_cluster_mean = sorted(cluster_mean_dict.items(), key=lambda x: x[1])
    
    df_cluster['Cluster'].replace({sorted_cluster_mean[0][0]: 'Cluster I', 
                                   sorted_cluster_mean[1][0]: 'Cluster II', 
                                   sorted_cluster_mean[2][0]: 'Cluster III', 
                                   sorted_cluster_mean[3][0]: 'Cluster IV'}, inplace=True)
    
    # Cluster column: categorical
    df_cluster['Cluster'] = pd.Categorical(df_cluster['Cluster'])
    
    # color palatte for stripplot
    cpalette = sns.color_palette(palette=None, n_colors=4)
    
    fig, ax = plt.subplots(1, figsize=(5.6118, 1.8))
    
    # plot stripplot
    sns.stripplot(x=df_cluster[metric][df_cluster.Cluster == "Cluster I"], label='Cluster I', color=cpalette[0], jitter=0.05, ax=ax)
    sns.stripplot(x=df_cluster[metric][df_cluster.Cluster == "Cluster II"], label='Cluster II', color=cpalette[1], jitter=0.05, ax=ax)
    sns.stripplot(x=df_cluster[metric][df_cluster.Cluster == "Cluster III"], label='Cluster III', color=cpalette[2], jitter=0.05, ax=ax)
    sns.stripplot(x=df_cluster[metric][df_cluster.Cluster == "Cluster IV"], label='Cluster IV', color=cpalette[3], jitter=0.05, ax=ax)
    
    ax.set_xlim(-0.04, 1.04)
    ax.legend(ncol=4, loc='lower center', handletextpad=0.1, edgecolor='black', framealpha=0.5)
    
    # annotate b)
    trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
    plt.text(0, 1, 'b)', verticalalignment='top', transform = ax.transAxes+trans)
    
    # save stripplot
    plt.tight_layout()
    plt.savefig(f'figures/Custering/{metric}/{metric}_stripplot.pdf', dpi=100)
    plt.close()
    
    # build table metric cluster performance and put on plot
    df_cluster_performance = pd.DataFrame(columns=['Number of points in cluster', 'Average correlation score in cluster', 'Average value of expected correlation'])

    for cluster in label_dict.values():
        cluster.replace('(', ', ')
        cluster = cluster[1:-1]
        cluster = cluster.split(', ')
        cluster = list(map(float, cluster))
        
        df_cluster_performance.loc[len(df_cluster_performance)] = cluster
    
    df_cluster_performance.sort_values('Average value of expected correlation', inplace=True)
        
    df_cluster_performance.index = ['Cluster I', 'Cluster II', 'Cluster III', 'Cluster IV']
    df_cluster_performance.index.name = metric 

    # save cluster performance dataframe
    df_cluster_performance.to_csv(f'figures/Custering/{metric}/{metric}_cluster_performance.csv')
```

### Reorganize data
table: left_id, right_id, all metrics

```python
left_ids = ['8', '9']
left_ids = ', '.join(left_ids)
right_ids = left_ids

sql = 'SELECT * FROM correlation_matrix WHERE left_id IN (%s) AND right_id IN (%s)' % (left_ids, right_ids)
df = pd.read_sql(sql, session.bind)
df
```

```python
def reorganize_matrix(session, left_entries, right_entries):
    """
    Function to transform the correlation_matrix in metacatalog to a wide format with columns
    left_id, right_id and all metric values.
    
    Parameters
    ----------
    session : sqlalchemy.orm.Session
        Session to the database.
    left_entries : list(metacatalog.models.Entry)
        List of entries for which the correlation matrix is to be loaded.
    right_entries : list(metacatalog.models.Entry)
        List of entries for which the correlation matrix is to be loaded.
    """
    # get left_id's
    left_ids = []
    for entry in left_entries:
        left_ids.append(str(entry.id))
    left_ids = ', '.join(left_ids) # remove brackets from list of ids to use them in the sql query below
    
    # get right_id's
    right_ids = []
    for entry in right_entries:
        right_ids.append(str(entry.id))
    right_ids = ', '.join(right_ids) # remove brackets from list of ids to use them in the sql query below
    
    
    # load correlation matrix for specified left_id's and right_id's
    sql = 'SELECT * FROM correlation_matrix WHERE left_id IN (%s) AND right_id IN (%s)' % (left_ids, right_ids)
    df = pd.read_sql(sql, session.bind)
    
    # get metrics for output column names
    sql = 'SELECT id, symbol FROM correlation_metrics'
    df_metrics = pd.read_sql(sql, session.bind)
    dict_metrics = dict(zip(df_metrics.id, df_metrics.symbol))
    
    # build result_dict, convert to pd.DataFrame later
    result_dict = {'left_id': [],
                   'right_id': []}
    
    # key for each metric
    for metric_id in sorted(df.metric_id.unique()):
        result_dict[dict_metrics[metric_id]] = []

    # fill dictionary
    for left_id in sorted(df.left_id.unique()):    
        for right_id in sorted(df.right_id.unique()):       
            result_dict['left_id'].append(left_id)            
            result_dict['right_id'].append(right_id)            
            for metric_id in sorted(df.metric_id.unique()):
                metric_value = df.loc[((df['left_id']==left_id) & (df['right_id']==right_id) & (df['metric_id']==metric_id))]['value']
                # add value if existing
                if metric_value.any():
                    result_dict[dict_metrics[metric_id]].append(float(metric_value))
                else:
                    result_dict[dict_metrics[metric_id]].append(np.nan)
    # create df from dict
    result_df = pd.DataFrame(result_dict)
    
    return result_df
```

Build function: 
- **input**: two datasets (left, right), metrics2compute
- **output**: timeseries plot, scatter plot, qq-plots, table with all metrics

```python
import pandas as pd
import numpy as np
from matplotlib.dates import DateFormatter, MonthLocator
import statsmodels.api as sm


def calculate_metrics_set(left, right, corr_metrics):
    """
    Calculate a set of correlation metrics for 1D timeseries DataFrames left and right.
    Also plots timeseries and scatter plot with linear regression.
    
    Parameters
    ----------
    left : pandas.DataFrame
        Left 1D timeseries DataFrame for which the metrics are calculated.
    right : pandas.DataFrame
        Right 1D timeseries DataFrame for which the metrics are calculated.
    corr_metrics : list[str]
        List of metrics to compute ('spearman', 'pearson').
        
    Returns
    -------
    df : pandas.DataFrame
        DataFrame with the values of the calculated metrics.
    """
    
    # append calculated metrics to data
    data = []
    
    if 'pearson' in corr_metrics:
        pearson_r = metrics.pearson_corr_coef(left.iloc[:,0], right.iloc[:,0])
        data.append({'metric': 'pearson_r', 
                     'value': round(pearson_r, 3)})
        
    if 'spearman' in corr_metrics:
        spearman_r = metrics.spearman_corr_coef(left.iloc[:,0], right.iloc[:,0])
        data.append({'metric': 'spearman_r', 
                     'value': round(spearman_r, 3)})
    
    # save metric results to df
    df = pd.DataFrame(data)
    
    # timeseries and scatter plot with linear regression & table with calculated metrics
    #fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,6))
    fig = plt.figure(figsize=(12,6))

    gs = fig.add_gridspec(2, 3)
    
    ax1 = fig.add_subplot(gs[0:2, 0])
    
    ax2 = fig.add_subplot(gs[0:2, 1])
    
    ax3 = fig.add_subplot(gs[0, 2])
    
    ax4 = fig.add_subplot(gs[1, 2])
    
    # timeseries plot
    # secondary y-axis
    ax1b = ax1.twinx()
    
    # plotting on the first y-axis
    ax1.plot('%s' % left.columns[0], data=left, alpha=1, color='slateblue')
    ax1.set(xlabel='Date',
            ylabel='%s' % left.columns[0], # not needed, scatterplot with same y-axis label
            title='Timeseries')
    
    # plotting on the second y-axis
    ax1b.set_ylabel('%s' % right.columns[0])
    ax1b.plot('%s' % right.columns[0], data=right, alpha=0.4, color='darkturquoise')
    
    # lines and labels from all axes for a single legend
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax1b.get_legend_handles_labels()
    ax1b.legend(lines + lines2, labels + labels2, loc='upper left');
    
    # format date labels
    plt.setp(ax1.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    
    # scatter plot with linear regression line    
    # linear regression
    x = left.iloc[:,0]
    y = right.iloc[:,0]
    coef = np.polyfit(x,y,1)
    poly1d_fn = np.poly1d(coef) 
    # poly1d_fn is now a function which takes in x and returns an estimate for y

    ax2.scatter(x, y, color='teal')
    ax2.plot(x, poly1d_fn(x), '--k')
    ax2.set(xlabel='%s' % left.columns[0],
            #ylabel='%s' % right.columns[0], # not needed -> timeseries plot with same y-label next to scatterplot
            title="Linearity?")
    
    
    # qq-plots for dat1 and dat2
    sm.qqplot(dat1.iloc[:,0], line='q', ax=ax3)
    ax3.set(title='Normality %s?' % dat1.columns[0])
    
    sm.qqplot(dat2.iloc[:,0], line='q', ax=ax4)
    ax4.set(title='Normality %s?' % dat2.columns[0])
    
    
    fig.tight_layout()
    
    plt.show()
    
    return df    
    
```
