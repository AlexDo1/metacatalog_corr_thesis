````python
import numpy as np
import skinfo.metrics as skinfo

right_df = api.find_entry(session, id=1064, include_partial=True)[0]
right_df = right_df.get_data()

left_df = api.find_entry(session, id=1014)[0]
left_df = left_df.get_data()

right = right_df['Cov(v\'Ts\')[(m*deg C)/s]']
left = left_df

harmonized_index = right[right.index.isin(left.index)].index
left = left.loc[harmonized_index]
right = right.loc[harmonized_index]

left = left.to_numpy()
right = right.to_numpy()

#left = np.hstack(left)
#right = np.hstack(right)

nan_indices = np.logical_not(np.logical_or(np.isnan(left), np.isnan(right)))

left = left[nan_indices]
right = right[nan_indices]

corr = skinfo.mutual_information(left, right, 'fd')

print(corr)
````

````python
x=left
y=right
bins='fd'

# jensen shannon

# get the bins
bins = np.histogram_bin_edges([x, y], bins)

# calculate unconditioned histograms
hist_x = np.histogram(x, bins=bins)[0]
hist_y = np.histogram(y, bins=bins)[0]

# calculate probabilities
px = (hist_x / np.sum(hist_x)) + 1e-15
py = (hist_y / np.sum(hist_y)) + 1e-15

# calculate m
pm = 0.5 * (px + py)

# cross entropy 
#px = px + 1e-15
#py = pm + 1e-15
#
#cross_xm = - px.dot(np.log2(py))
#
#px = py + 1e-15
#py = pm + 1e-15
#
#cross_ym = - px.dot(np.log2(py))
#
## entropy x
#p = px + 1e-15
#entropy_x = - p.dot(np.log2(p))
#
#p = py + 1e-15
#entropy_y = - p.dot(np.log2(p))
#
#
#kl_xm = cross_xm - entropy_x
#kl_ym = cross_ym - entropy_x
````

````python
import skinfo.metrics as skinfo

pm = 0.5 * (px + py)

# calculate kullback-leibler divergence between px and pm & py and pm
kl_xm = skinfo.kullback_leibler(px, pm, bins=bins, use_probs=True)
kl_ym = skinfo.kullback_leibler(py, pm, bins=bins, use_probs=True)
````
